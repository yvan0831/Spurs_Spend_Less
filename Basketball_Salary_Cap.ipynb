```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = '/content/drive/MyDrive/Coding Dojo/Week 3/NBA_Salary_History.xlsx'
df = pd.read_excel(data)
df.head()
```

```python
df_lakers = df[(df['Team'] == 'Los Angeles Lakers')]
df_lakers.head(10)
```

```python
df_spurs = df[(df['Team'] == 'San Antonio Spurs')]
df_spurs.head()
```

```python
plt.figure(figsize = (19,10))
plt.plot(df_lakers['Season'], df_lakers['Total Salary'], c = 'purple', label = 'Lakers');
plt.plot(df_spurs['Season'], df_spurs['Total Salary'], c = 'grey', label = 'Spurs');
plt.scatter(df_lakers['Season'], df_lakers['Total Salary'], c = 'purple',);
plt.scatter(df_spurs['Season'], df_spurs['Total Salary'], c = 'grey');
plt.title('Spurs Winning and Spending Less', fontsize = 25);
plt.xticks(rotation = 45);
plt.ylim(bottom = 0, top = 120000000);
plt.xlabel('Season', fontsize = 20)
plt.ylabel('Total Salary ($)', fontsize = 20);
plt.scatter(8, 43681143, c = 'grey', marker = '*', s = 300, label = 'Spurs Championship');
plt.scatter(12, 53182559, c = 'grey', marker = '*', s = 300);
plt.scatter(15, 63570274, c = 'grey', marker = '*', s = 300);
plt.scatter(23, 63858408, c = 'grey', marker = '*', s = 300);
plt.scatter(14, 47504123, c = 'grey', marker = '*', s = 300);
plt.scatter(9, 55018533, c = 'purple', marker = '*', s = 300, label = 'Lakers Championship');
plt.scatter(10, 58968213, c = 'purple', marker = '*', s = 300);
plt.scatter(11, 54262017, c = 'purple', marker = '*', s = 300);
plt.scatter(18, 78245793, c = 'purple', marker = '*', s = 300);
plt.scatter(19, 91378064, c = 'purple', marker = '*', s = 300);
plt.legend(loc = (0.7,0.28), prop={'size' : 20});
```
